```python
#Jack Schwyn
#Lab 03
import numpy as np #Import numpy for arrays
import pandas as pd #Import pandas for dataframes
import matplotlib.pyplot as plt #Import matplotlib.pyplot for data visualization
import seaborn as sns #Import seaborn for more data visulizations.
```

```python
#Import Data
netflixdata = pd.read_csv('netflix_titles.csv') #Here, we import and "read" the data as a pandas dataframe.
netflixdata.dropna().iloc[:10] #Dropna deletes rows with "NA"s.
#Iloc[10] displays the first ten rows in the dataframe
```

```python
#Data Exploration (Movies)
#Here, we sort the data based on type, in this case Movies. 
movies = netflixdata[netflixdata.type=='Movie'] #This dataframe only shows movies.
movies.dropna() #Get rid of NAs.
```

```python
#Data Exploration (TV shows)
#Similar to the movies, we are sorting by type.
tvshows = netflixdata[netflixdata.type=='TV Show'] #This dataframe only shows movies.
tvshows.dropna() #Get rid of "NA"s.
```

```python
#Pie Chart
netflixdata = netflixdata.dropna()
df = netflixdata['type'].value_counts() #Here, we are counting each time there is a movie, and each time there is a tv show.
#This pie chart is based on counting the movies and tv shows.
df.plot.pie(autopct='%.2f%%', labels = ['Movies', 'TV Shows']) #PLot the pie chart, assign labels, and format the percentages.
plt.title("Percetnage of Movies and TV Shows on Netflix") #Pie Chart title.
plt.legend(loc = 'upper right') #Pie chart legend for extra clarity.
plt.ylabel(" ")
plt.show() #Show plot.
df
```

```python
#Bar Graph (TV Shows)
tvshows = netflixdata[netflixdata.type=='TV Show'] #Sort based on TV Shows
tvshowsna = tvshows.dropna() #Remove "NA"s
df = (tvshowsna['country'].sort_values().value_counts(sort = False)) 
#Go through dataframe and count the number of TV Shows each country has.
color = ['red', 'blue', 'grey', 'green', 'pink', 'orange', 'black', 'lightblue', 'purple', 'lightgreen']
#Assign colors to make the chart more clear and visually appealing.
barchart = (tvshowsna['country'].value_counts()).iloc[:10] #Show the top ten.
barchart.plot(kind='bar', color = color) #Plot bar plot.
plt.title("Countries with the Most TV Shows") #PLot title.
plt.ylabel('Number of Shows') #Label teh y axis.
plt.grid() #Plot the grid.
plt.show() #Show the plot.

df #Present the "raw data," or what the bar plot is based on.
```

Most of Netflix's shows are distributed exclusively in the United States and the United Kingdon. This indicates that Netflix caters to mostly English speaking countries, which makes sense since Netflix is an American company. However, if they want to appeal to international audiences, for example Asian audiences, they should aqurie more K-Dramas and Anime. Not only are these popular in their respective countries, but they are popular in the U.S. as well.

```python
#Rating Analysis (TV Shows)
tvshows = tvshows.dropna()
df = (tvshows['rating'].sort_values().value_counts(sort = False))
#Go through dataframe and count how many times a certain rating appears.
sns.countplot(y="rating", data = tvshows, order=tvshows["rating"].value_counts(ascending=False).index)
#Here, we are using a sns.countplot to present the data. 
#The x axis would be the rating, and the tvshowsna would be the data we are using. 
#We are sorting by descending order, which is why ascending = False.
plt.title("TV Show Ratings") #PLot title.
plt.ylabel("Number of TV Shows") #Label y axis.
plt.grid() #Show grid. 
plt.show() #Show plot.

df #Present "raw data" or what the bar plot is based on.
```

```python
#I believe that a bar chart would be appropriate for this data set because "Seasons" is categorial. (Duration)
#I did make a histogram as well though.
tvshows = netflixdata[netflixdata.type=='TV Show'] #Sort by type.
tvshowsna = tvshows.dropna() #Remove "NA"s.
df = (tvshowsna['duration'].sort_values().value_counts(sort = False))
#Count how may times a certain show duration appears (for "raw data".
barcharttv = (tvshowsna['duration'].value_counts())
#Count how may times a certain show duration appears for the plot.
barcharttv.plot(kind='bar') #Plot bar plot.
plt.title("Duration of TV Shows") #Plot title
plt.ylabel("Number of TV Shows") #Label y axis
plt.grid() #Show grid.
plt.show() #Show plot. 

df #Show "raw data" or what the plot is based on.
```

Most TV shows on Netflix run for 1 season before eventually being cancelled. This shows that, if a show is not getting enough engagement, Netflix will
cancel it as they do not want to fund anything that does not generate engagement. Movies tend to be around 90 to 120 minutes long, but TV Shows tend to have several 40 minute episodes per season. 20 minutes episodes if it is a sit-com. If the show is not engaging within the first episode or season, the viewer might want to waste their time watching more 40 minute (or 20 minute) episodes. The histogram will also show you that most shows on Netflix run for 1 season.

```python
#Histogram
tvshows = netflixdata[netflixdata.type=='TV Show'] #Sort by type.
df = tvshows.dropna() #Remove "NA"s.
array = df[['duration']].to_numpy() #Create an array based on the duration.
plt.hist(array) #plot the histogram.
plt.title("Distribution of TV Show Durations") #PLot title.
plt.ylabel("Frequency") #Plot y axis.
plt.grid() #Show grid.
plt.xticks(rotation=90) #Rotate the x tick to make the graph visually appealing.
plt.show() #Show plot.
```
