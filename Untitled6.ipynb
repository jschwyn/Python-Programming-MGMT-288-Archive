```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emily'],
    'Age': [25, 30, 27, 29, 26],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'San Francosco']
}

df = pd.DataFrame(data)
print(f"Original Dataframe \n (df) \n")
df.loc[0]
print(df.loc[:,'Name'])
```
