```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
flower_df = sns.load_dataset("iris")
flower_df
```

```python
setosa_df = flower_df[flower_df.species == 'setosa']
versicolor_df = flower_df[flower_df.species == 'versicolor']
plt.hist(setosa_df.sepal_width, alpha = 0.4, bins = np.arange(2, 5, 0.25), edgecolor = 'white')
plt.hist(versicolor_df.sepal_width, alpha = 0.4, bins = np.arange(2, 5, 0.25), edgecolor = 'white')
```
